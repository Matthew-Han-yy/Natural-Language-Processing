```python
# big_list = [small_list]
big_list = []

for page in range(1,10+1):
    url = "https://www.internsg.com/page/{}/?s=data+scientist".format(page)
    
    driver.get(url)
    sleep(1)
    print "Accessing... ", url
    html = driver.page_source
    html = BeautifulSoup(html, 'lxml')
    
    entry_urls = []
    for post in html.find_all('div', {'class':'post-footer'}):
        for site in post.find_all('a', href=True):
            entry_urls.append(site['href'])
    
    for index, i in enumerate(entry_urls):

        driver_next = webdriver.Chrome(chromedriver)
        driver_next.get(i)
        sleep(1)
        html_next = driver_next.page_source
        html_next = BeautifulSoup(html_next, 'lxml')
        driver_next.close()

        # small_list = ['Company', 'Role', 'Salary', 'Job Description']
        small_list = []

        print "Getting Link {}".format(index)

        company = html_next.find_all('dd')[0].text
        small_list.append(company)

        role = html_next.find_all('dd')[1].text
        small_list.append(role)

        salary = html_next.find_all('dd')[6].text
        small_list.append(salary)

        jd = html_next.find_all(class_="dl-horizontal odd")[-1].text
        small_list.append(jd)

        big_list.append(small_list)
    
driver.close()

        
print big_list
print len(big_list)
    
```

```python
import pandas as pd

df = pd.DataFrame(big_list, columns=['company', 'role', 'salary', 'JD'])
df.head(20)
```

```python
#Function to clean JD. 
def clean_jd(x):
    # Drop "Job Description" title.
    x = x.strip('\nJob Description\n\n')
    return x

# Apply to JD column
df['JD'] = df['JD'].apply(lambda x: clean_jd(x))

# Apply clean non-ascii.
df['JD'] = df['JD'].apply(lambda x: ''.join([i if ord(i) < 128 else ' ' for i in x]))

df.head(20)
```
